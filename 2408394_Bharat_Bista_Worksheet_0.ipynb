Task - 1:
Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

```python
def unit_converter(choice, value):
    """
    Performs unit conversion based on the selected option.

    Parameters:
    choice (int): The conversion option selected by the user.
    value (float): The numeric value that needs to be converted.

    Returns:
    float: Converted value based on the selected conversion.
    """

    if choice == 1:
        return value * 3.28084          # meters change to  feet
    elif choice == 2:
        return value / 3.28084          # feet change to meters
    elif choice == 3:
        return value * 2.20462          # kg change to pounds
    elif choice == 4:
        return value / 2.20462          # pounds change to kg
    elif choice == 5:
        return value * 0.264172         # liters change to gallons
    elif choice == 6:
        return value / 0.264172         # gallons change to liters
    else:
        return None


print("Unit Conversion Program")
print("-----------------------")
print("1. Meters to Feet")
print("2. Feet to Meters")
print("3. Kilograms to Pounds")
print("4. Pounds to Kilograms")
print("5. Liters to Gallons")
print("6. Gallons to Liters")

try:
    choice = int(input("Select conversion type (1-6): "))
    value = float(input("Enter the value to convert: "))

    result = unit_converter(choice, value)

    if result is None:
        print("Unsupported conversion type.")
    else:
        print("Converted value:", result)

except ValueError:
    print("Error: Please enter valid numeric input.")
```

```python
def find_sum(numbers):
    """
    Returns the sum of all numbers in the list.
    numbers: list of numbers
    return: sum of numbers
    """
    return sum(numbers)


def find_average(numbers):
    """
    Returns the average of numbers in the list.
    numbers: list of numbers
    return: average value
    """
    return sum(numbers) / len(numbers)


def find_max(numbers):
    """
    Returns the maximum number in the list.
    numbers: list of numbers
    return: maximum value
    """
    return max(numbers)


def find_min(numbers):
    """
    Returns the minimum number in the list.
    numbers: list of numbers
    return: minimum value
    """
    return min(numbers)


print("Choose Operation")
print("1. Sum")
print("2. Average")
print("3. Maximum")
print("4. Minimum")

try:
    choice = int(input("Enter your choice (1-4): "))

    nums = input("Enter numbers separated by space: ")
    numbers = [float(x) for x in nums.split()]

    if len(numbers) == 0:
        print("List is empty")

    elif choice == 1:
        print("Sum =", find_sum(numbers))

    elif choice == 2:
        print("Average =", find_average(numbers))

    elif choice == 3:
        print("Maximum =", find_max(numbers))

    elif choice == 4:
        print("Minimum =", find_min(numbers))

    else:
        print("Invalid choice")

except ValueError:
    print("Invalid input. Please enter numbers only.")
```

#4.3 Exercise on Nested List:

1. Flatten a Nested List:
Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
    lst (list): A nested list containing sublists.

    Returns:
    list: A flattened list with all elements in one dimension.
    """
    flat_list = []

    for sublist in lst:

        for item in sublist:
            flat_list.append(item)

    return flat_list


# Example
nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", result)
```

2. Accessing Nested List Elements:
Write a Python function that extracts a specific element from a nested list given its indices.


```python
def access_nested_element(lst, indices):
    """
    Returns a specific element from a nested list using a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list): A list of indices indicating the position of the element.

    Returns:
    element: The value located at the specified nested position.
    """

    element = lst

    for index in indices:
        element = element[index]

    return element


# Example
lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]

result = access_nested_element(lst, indices)

print("Nested List:", lst)
print("Indices:", indices)
print("Accessed Element:", result)
```

3. Sum of All Elements in a Nested List:
Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
def sum_nested(lst):
    """
    Calculate the sum of all numbers in a nested list.

    Parameters:
    lst (list): A nested list that may contain numbers or other lists.

    Returns:
    int/float: Sum of all numbers in the nested list.
    """
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)   # recursive call for sublist
        else:
            total += item

    return total


# Example
numbers = [[1, 2], [3, [4, 5]], 6]

result = sum_nested(numbers)

print("Nested List:", numbers)
print("Sum of all elements:", result)
```

4. Remove Specific Element from a Nested List:
Write a Python function that removes all occurrences of a specific element from a nested list.

```python
def remove_element(lst, elem):
    """
    Remove all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): A nested list that may contain other lists.
    elem: The element to remove.

    Returns:
    list: Modified list with the element removed.
    """
    result = []

    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))  # process sublist
        elif item != elem:
            result.append(item)

    return result


# Example
lst = [[1, 2], [3, 2], [4, 5]]
elem = 2

result = remove_element(lst, elem)

print("Original Nested List:", lst)
print("Element to remove:", elem)
print("Modified List:", result)
```

5. Find the Maximum Element in a Nested List:
Write a Python function that finds the maximum element in a nested list (regardless of depth)

```python
def find_max(lst):
    """
    Find the maximum element in a nested list.

    Parameters:
    lst (list): A nested list that may contain numbers or other lists.

    Returns:
    int/float: The maximum element in the nested list.
    """
    maximum = None

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)  # recursive call
        else:
            value = item

        if maximum is None or value > maximum:
            maximum = value

    return maximum


# Example
numbers = [[1, 2], [3, [4, 5]], 6]

result = find_max(numbers)

print("Nested List:", numbers)
print("Maximum element in the nested list:", result)
```

6. Count Occurrences of an Element in a Nested List:
Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
    lst (list): The nested list.
    elem: The element to count.

    Returns:
    int: Number of times the element appears.
    """

    count = 0

    for sublist in lst:
        for item in sublist:
            if item == elem:
                count += 1

    return count


# Example
lst = [[1, 2], [2, 3], [2, 4]]
elem = 2

result = count_occurrences(lst, elem)

print("Nested List:", lst)
print("Element to count:", elem)
print("Occurrences:", result)
```

7. Flatten a List of Lists of Lists:
Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list regardless of depth.

    Parameters:
    lst (list): A deeply nested list.

    Returns:
    list: A fully flattened list.
    """

    flat_list = []

    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # recursive call
        else:
            flat_list.append(item)

    return flat_list


# Example
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

result = deep_flatten(nested_list)

print("Original nested list:", nested_list)
print("Flattened list:", result)
```

8. Nested List Average:
Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
    lst (list): A nested list containing numbers.

    Returns:
    float: The average of all elements.
    """

    total = 0
    count = 0

    for sublist in lst:
        for item in sublist:
            total += item
            count += 1

    return total / count


# Example
nested_list = [[1, 2], [3, 4], [5, 6]]

result = average_nested(nested_list)

print("Nested List:", nested_list)
print("Average:", result)
```
